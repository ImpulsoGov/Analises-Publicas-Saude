```python
#bibliotecas
import pandas as pd
pd.set_option('display.max_columns', None)

# scripts
import treat_functions as tf
```

## Respostas do formulário

```python
# percentual de preenchimento
tf.percentualcol(forms)
```

```python
forms
```

```python
# coloca strings em caixa alta e depois as normaliza
forms = tf.touppercase(forms)
forms = tf.normalizestrings(forms)

# normaliza cpf
forms['CPF'] = tf.normalizecpf(forms['CPF'])

# normaliza data de nascimento para o formato %Y/%m/%d
forms['Data de nascimento'] = tf.normalizedata(forms['Data de nascimento'], '%Y-%m-%d')
```

```python
forms.head()
```

```python
duplicated = forms[forms.duplicated(subset=['CPF','Nome completo'], keep='first')]
duplicated['CPF'].count()
```

```python
forms.shape[0]
```

```python
# remoção das duplicatas
forms_ = tf.removeduplicatas(forms,'Timestamp',['CPF','Nome completo'])
forms_.shape[0]
```

```python
forms_.to_csv('data/formulario_deduplicado_20201209.csv')
```
