### Comparação das versões

```python
v2_categories = {0: "novo normal", 1: "moderado", 2: "alto", 3: "altíssimo"}
```

```python
change_alert = (v1["cities_farol"]
    .set_index("city_id")[["overall_alert"]]
    .join(v2["cities_farol"].set_index("city_id")["overall_alert"],
          lsuffix="_v1",
          rsuffix="_v2")
    .assign(overall_alert_v2 = lambda df: df["overall_alert_v2"].map(v2_categories).fillna("sem classificação"))
    .assign(overall_alert_v1 = lambda df: df["overall_alert_v1"].fillna("sem classificação")))

change_alert.reset_index().groupby(["overall_alert_v1", "overall_alert_v2"]).count()
```

```python
(change_alert
.reset_index()
.groupby(["overall_alert_v1", "overall_alert_v2"])
.count()
.sort_values(by="city_id", ascending=False))
```

```python
def gen_classification_table(df, subset):
    cols = [i for i in df.columns if "classification" in i]
    return df[df["city_id"].isin(subset)][cols].apply(pd.Series.value_counts)

def gen_indicators_table(df, subset, version):
    if version == "v1":
        cols = ["dday_beds_best", "rt_10days_ago_avg", "subnotification_rate"]
    if version == "v2":
        cols = ["dday_icu_beds", "rt_most_likely", "daily_cases_mavg_100k", "subnotification_rate"]
        
    if len(df[df["city_id"].isin(subset)]) < 5:
        return df[df["city_id"].isin(subset)][cols]
    else:
        return df[df["city_id"].isin(subset)][cols].describe()
```
